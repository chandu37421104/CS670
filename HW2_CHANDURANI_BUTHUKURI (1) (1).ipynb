PyTorch is a widely acclaimed open-source machine learning library for Python, famous for its flexibility, user-friendly interface, and dynamic computation graph. Computations on Tensors:

Tensors are the core component of PyTorch, serving as the building blocks for model architecture and data. They are similar to arrays and matrices, enabling efficient computation and manipulation of multi-dimensional data. Here's a brief introduction to performing basic tensor

```python
import torch

# Create tensors
x = torch.tensor([1, 2, 3])
y = torch.tensor([4, 5, 6])

# Basic operations
sum_xy = x + y
product_xy = x * y

print("Sum:", sum_xy)
print("Product:", product_xy)
```

**Potential use cases for such operations in data preprocessing.**

1. These operations, like making new data frames and combining and removing the required columns from a given CSV file, help to create our own data set.

2. In ML training, we often train the model in batches. Combining features and labels into a single tensor can simplify creating batches.

3. During ML model training, we often normalize or standardize the data. Having all features and labels in single tensor makes the process less complicated in certain cases.

4. It's important to note that while combining features and labels into a single tensor can be convenient for certain operations, it's not always necessary. The decision to combine them should be based on the specific requirements Of data processing pipeline and the constraints of the ML model.

Task 7: Loading MNIST dataset

Load MINST data set with the code below:

```python
from torchvision import datasets, transforms

# Define a transform to normalize the data
transform = transforms.Compose([transforms.ToTensor(),
                                transforms.Normalize((0.5,), (0.5,))])

# Download and load the training data
trainset = datasets.MNIST('~/.pytorch/MNIST_data/', download=True, train=True, transform=transform)
trainloader = torch.utils.data.DataLoader(trainset, batch_size=64, shuffle=True)

# Get one batch of training images and labels
images, labels = next(iter(trainloader))
print("Shape of images:", images.shape)
print("Shape of labels:", labels.shape)
```

Todo: Print the shape of the images and labels tensors.

```python
#todo
for images, labels in trainloader:
    print("Shape of images:", images.shape)
    print("Shape of labels:", labels.shape)
```

Todo: Extract and visualize the first 5 images from the batch.

```python
#todo
import matplotlib.pyplot as plt
figures, axes = plt.subplots(1, 5, figsize=(15, 3))
for i in range(5):
    ax = axes[i]
    img = images[i].squeeze()  # Remove the channel dimension
    ax.imshow(img, cmap='gray')
    ax.set_title(f'Label: {labels[i].item()}')
    ax.axis('off')

plt.show()
```
